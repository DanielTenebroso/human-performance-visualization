# ðŸ§  Human Performance Data Visualization

This notebook presents three visualizations designed to analyze and communicate five key dimensions of human performance:

- **ðŸ“Š Radar Chart** â€“ compares multiple performance metrics in a single view.
- **ðŸ“ˆ Line Chart** â€“ shows the weekly progression of cognitive clarity over time.
- **ðŸ§  Concept Map** â€“ represents how the core dimensions of performance relate to a central concept.

The metrics included are:
- Physical Endurance
- Cognitive Clarity
- Emotional Balance
- Detox Level
- Longevity Potential

The goal is to showcase data visualization skills using Python, with libraries such as `matplotlib`, `pandas`, and `graphviz`, applied to topics of personal optimization and human development.

```python
import matplotlib.pyplot as plt
import numpy as np

labels = [
    "Physical Endurance",
    "Cognitive Clarity",
    "Emotional Balance",
    "Detox Level",
    "Longevity Potential"
]
values = [7.5, 8.2, 6.8, 7.0, 9.0]
values += values[:1]
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.plot(angles, values, color='crimson', linewidth=2)
ax.fill(angles, values, color='lightcoral', alpha=0.5)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_yticks([2, 4, 6, 8, 10])
ax.set_ylim(0, 10)
plt.title("ðŸ“Š Human Performance Radar Chart")
plt.show()
```

```python
import pandas as pd

weeks = ["Week 1", "Week 2", "Week 3", "Week 4"]
cognitive_scores = [6.5, 7.0, 8.0, 8.2]

df = pd.DataFrame({"Week": weeks, "Cognitive Clarity": cognitive_scores})

# GrÃ¡fico de lÃ­nea
plt.figure(figsize=(8, 4))
plt.plot(df["Week"], df["Cognitive Clarity"], marker='o', color='royalblue', linewidth=2)
plt.title("ðŸ“ˆ Cognitive Clarity Over Time", fontsize=14)
plt.xlabel("Week")
plt.ylabel("Score")
plt.grid(True)
plt.ylim(0, 10)
plt.show()
```

```python
df.plot(kind="bar", x="Week", y="Cognitive Clarity", color='skyblue', legend=False)
plt.title("ðŸ“Š Weekly Cognitive Clarity")
plt.ylim(0, 10)
plt.ylabel("Score")
plt.show()
```
